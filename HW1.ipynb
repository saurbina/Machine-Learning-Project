```python
#Import python libraries
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib as mpl
import seaborn as sns
```

```python
#Read csv file mtcars
url = "https://vincentarelbundock.github.io/Rdatasets/csv/datasets/mtcars.csv"
mtcars = pd.read_csv(url)
```

### 2.  Display the first five rows of the data.

```python
# List first 5 records
mtcars.head(5)
```

```python
#List of the last 10 records
mtcars.tail(10)
```

### 3.  Calculate the average of the mpg column for all cars within each category of the cyl column.



```python
# Average mpg by cyl
mtcars.groupby("cyl")[["mpg"]].mean()
```

### 4.  Create a histogram using the mpg column

```python
# Import pyplot file from matplotlib
from matplotlib import pyplot as plt
```

```python
#Histogram of mpg
mtcars["mpg"].hist()
plt.xlabel("mpg")
plt.ylabel("Count")
plt.title("MPG Histogram")
plt.show()
```

### 5. Choose two variables in the data and create a scatterplot.



```python
#Scatterplot mpg and disp
plt.scatter(mtcars["mpg"],mtcars["disp"])
plt.xlabel("mpg")
plt.ylabel("disp")
plt.title("mpg by disp")
plt.show()
```
